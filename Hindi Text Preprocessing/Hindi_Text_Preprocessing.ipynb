```python
# Hindi Text Processing Project

# Step 1: Import libraries
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Step 2: Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')

# Step 3: Load Hindi text file
with open("hindi_text.txt", "r", encoding="utf-8") as f:
    hindi_text = f.read()

# Step 4: Tokenize Hindi text
tokens = nltk.word_tokenize(hindi_text)

# Step 5: Define some sample Hindi stopwords
hindi_stopwords = ["यह", "और", "था", "थे", "है", "की", "के", "को", "में", "पर"]

# Step 6: Remove stopwords
filtered_tokens = [word for word in tokens if word not in hindi_stopwords]

# Step 7: Join tokens back to cleaned text
hindi_text_clean = " ".join(filtered_tokens)

# Step 8: Generate WordCloud (using Hindi font)
wordcloud = WordCloud(
    font_path="lohit-devanagari.ttf",   # keep this file in same folder
    width=800, height=400,
    background_color="white"
).generate(hindi_text_clean)

# Step 9: Show the WordCloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
