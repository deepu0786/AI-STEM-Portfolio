# 📊 COVID-19 Data Analysis (India)

This project demonstrates basic **data analysis and visualization** using Python, Pandas, and Matplotlib.  
Dataset: `covid_india.csv` (sample dataset for India’s COVID cases, deaths, and vaccinations).  

🔹 Goals:
- Load and explore the dataset  
- Visualize daily new cases, deaths, and vaccination progress  
- Extract useful insights  

This project is part of my **AI & STEM Portfolio**.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 🔹 Step 1: Load the Dataset
We start by loading the dataset (`covid_india.csv`) into a Pandas DataFrame.


```python
df = pd.read_csv("covid_india.csv")
df.head()
```

## 🔹 Step 2: Explore the Dataset
Check basic info, data types, and summary statistics.


```python
print("Dataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

## 📈 Step 3: Daily New Cases
Let’s plot the number of new cases reported each day.


```python
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['new_cases'], marker='o')
plt.title("Daily New COVID Cases in India")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 💀 Step 4: Daily Deaths
Now we plot the number of daily deaths.


```python
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['new_deaths'], color='red', marker='o')
plt.title("Daily COVID Deaths in India")
plt.xlabel("Date")
plt.ylabel("New Deaths")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 💉 Step 5: Vaccination Progress
Finally, let’s see how vaccinations progressed over time.


```python
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['total_vaccinations'], color='green', marker='o')
plt.title("Total Vaccinations in India")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 🔍 Step 6: Key Insights
From the analysis, we can summarize:


```python
print("\n--- Insights ---")
print("1. Peak cases recorded:", df['new_cases'].max())
print("2. Peak deaths recorded:", df['new_deaths'].max())
print("3. Total vaccinations achieved:", df['total_vaccinations'].max())
```
